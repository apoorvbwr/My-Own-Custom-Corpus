```python
from PyPDF4 import PdfFileReader
```

```python
def extract_pdf(pdf_filename, password = ' '):
    pdf_file = open(pdf_filename,'rb')
    read_pdf = PdfFileReader(pdf_file)
    if password != ' ':
        read_pdf.decrypt(password)
    pdf_text = []
    for i in range(read_pdf.getNumPages()-1):
        pdf_text.append(read_pdf.getPage(i).extractText())
    return "\n".join(pdf_text)
        
```

```python
def extract_word_text(word_file_name):
    doc = docx.Document(word_file_name)
    word_text = []
    for para in doc.paragraphs:
        word_text.append(para.text)
    return '\n'.join(word_text)
```

```python
TF_IDF_lists = []
for doc_tokens in document_filtered_tokens:
    doc_tokens_array = np.array(doc_tokens)
    tf_idf_list = []
    for token in vocabulary:
        if token in doc_tokens:
            tf = np.count_nonzero(doc_tokens_array == token) / len(doc_tokens)
            idf = math.log(number_of_documents/tf)
            tf_idf_list.append(tf*idf)
        else:
            tf_idf_list.append(0)
    TF_IDF_lists.append(tf_idf_list)
    
```

```python
tag_mapping_dict = {'DET':'D', 'NOUN':'N', 'ADJ':'J', 'VERB':'V', 'ADP':'P', '.':'C', 'PRON':'R', 'NUM':'U', 'X':'O'}
```

```python
portion_of_training_data = []
```

```python
def training_sents(one_sentence):
    tags = [tag for token,tag in one_sentence]
    tokens = [token for token,tag in one_sentence]
    blanklist = []
    for t in tags:
        if t not in tag_mapping_dict.keys():
            blanklist.append('N')
        else:
            blanklist.append(tag_mapping_dict[t])
    return (" ".join(tokens),{'tags':blanklist})
```

```python
for tagged_sentence in TEXTS:
    
    portion_of_training_data.append(training_sents(tagged_sentence))
```

```python
json_format_training_data_portion = []
for example in portion_of_training_data:
    blank_dict = dict()
    blank_dict['text'] = example[0]
    blank_dict['label'] = example[1]['tags']
    
    #blank_dict['answer'] = "accept"
    json_format_training_data_portion.append(blank_dict)
```

```python
import numpy as np
```

```python
bootstrapped_examples = []
for example in portion_of_training_data:
    for i in range(0,20):
        l = len(example[1]['tags'])
        a = np.array(list(tag_mapping_dict.values()))
        wrong_tags = list(np.random.choice(a,size=l))
        bootstrapped_examples.append(tuple([example[0],{'tags':wrong_tags}]))
```

```python
json_format_bootstrap = []
for example in bootstrapped_examples:
    blank_dict = dict()
    blank_dict['text'] = example[0]
    blank_dict['label'] = example[1]['tags']
    
    #blank_dict['answer'] = "reject"
    json_format_bootstrap.append(blank_dict)
```

```python
bootstrap_training_examples = json_format_training_data_portion + json_format_bootstrap
```
